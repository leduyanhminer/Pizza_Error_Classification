```python
import pandas as pd
```

```python
df = pd.read_json('data/pizzacam.json')
df.head()
```

```python
id_list = [id_item['$oid'] for id_item in df['_id']]
image_url_list = [image_url_item for image_url_item in df['image_url']]
```

```python
all_error_list = []
for item in df['error_list']:
    if type(item) is dict:
        for category, errors in item.items():
            for error in errors:
                if error not in all_error_list:
                    all_error_list.append(error)
```

```python
for error in all_error_list:
    df[error] = 0
```

```python
for index, row in df.iterrows():
    if type(row['error_list']) is not float:
        data = row['error_list']
        if isinstance(data, dict):
            for key, value in data.items():
                for element in value:
                    if element in all_error_list:
                        df.at[index, element] = 1
```

```python
df['id'] = [id_item['$oid'] for id_item in df['_id']]
```

```python
df = df.drop('_id', axis=1)
```
